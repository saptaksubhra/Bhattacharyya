A positive slope moves in the upward direction or is upward sloping. Negative Slope: A slope in which two variables, i.e., variable at x-axis and variable at the y-axis, shows a negative relationship is known as negative slope.
With positive slope the line moves upward when going from left to right. With negative slope the line moves down when going from left to right.
Ans when the slope becomes zero then the line becomes parallel to x axis.

```python
#Example of a positive slope
# The line is sloping upwards to the right.
try:
    import matplotlib.pyplot as plt
    x = [0,1,2,3,4]
    y = [1,3,5,7,9]
    plt.plot(x,y)
    plt.show()
except:
    print('positive slope')
```

```python
#Example of a negative slope
# The line is sloping downwards to the right.
try:
    import matplotlib.pyplot as plt
    x = [10,8,6,4,2]
    y = [2,4,6,8,10]
    plt.plot(x,y)
    plt.show()
except:
    print('negative slope')
```

6.Use the formulas for a and b to explain ordinary least squares.

In statistics, ordinary least squares (OLS) is a type of linear least squares method for estimating the unknown parameters in a linear regression model. OLS chooses the parameters of a linear function of a set of explanatory variables by the principle of least squares: minimizing the sum of the squares of the differences between the observed dependent variable (values of the variable being observed) in the given dataset and those predicted by the linear function of the independent variable.

10.Describe the regression analysis assumptions and the BLUE principle.

The regression analysis assumptions:
Linear relationship.
Multivariate normality.
No or little multicollinearity.
No auto-correlation.
Homoscedasticity.

BLUE principle:
Under 1 - 6 (the classical linear model assumptions) OLS is BLUE (best linear unbiased estimator), best in the sense of lowest variance. 
Under 1 - 5 (the Gauss-Markov assumptions) OLS is BLUE and efficient.
Under 1 - 4, OLS is unbiased, and consistent.
