1.Find the range RX of the random variable X.
2.Find P(X ≤ 0.5)
3.Find P(0.25<X<0.75)
4.P(X = 0.2|X<0.6)

1.The range of X can be found from the PMF. The range of X consists of possible values for X. Here we have
RX={0.2,0.4,0.5,0.8,1}.

2.The event X≤0.5 can happen only if X is 0.2,0.4, or 0.5. Thus,
P(X≤0.5)	=P(X∈{0.2,0.4,0.5})
=P(X=0.2)+P(X=0.4)+P(X=0.5)
=PX(0.2)+PX(0.4)+PX(0.5)
=0.1+0.2+0.2=0.5

3.Similarly, we have
P(0.25<X<0.75)	=P(X∈{0.4,0.5})
=P(X=0.4)+P(X=0.5)
=PX(0.4)+PX(0.5)
=0.2+0.2=0.4

This is a conditional probability problem, so we can use our formula P(A|B)=P(A∩B)P(B).
We have
P(X=0.2|X<0.6)	=P((X=0.2) and (X<0.6)) / P(X<0.6)
=P(X=0.2) / P(X<0.6)
=PX(0.2) / PX(0.2)+PX(0.4)+PX(0.5)
=0.1 / (0.1+0.2+0.2)=0.2
