Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). This data structure can be converted into NumPy ndarray with the help of Dataframe.to_numpy() method.
Syntax: Dataframe.to_numpy(dtype = None, copy = False)
Parameters:
dtype: Data type which we are passing like str.
copy: [bool, default False] Ensures that the returned value is a not a view on another array.
Returns:
numpy.ndarray
Now we are changing the Pandas DataFrame into numpy array by using a method DataFrame.to_numpy(). But we should always clean the data first to have the accurate result and in this method only we are using the first five values of any column by using.head() method.

```python
try:
    import pandas as pd
    df = pd.read_csv("cricket.csv")
    df.dropna(inplace = True)
    bravo = pd.DataFrame(df['Player'].head())
    print(bravo.to_numpy())
    print(type(bravo.to_numpy()))
    
except:
    print('conversion of pd df to numpy array')
```

Q3. Identify some of the plotting techniques that are used to produce a stock-market chart.

The plotting techniques that are used to produce a stock-market chart is by data visualization. Bokeh is a Python interactive data visualization that can be used to visualize stock market data. Visualization is done by the plotiing module. first of all we need to import bokeh to download the sample datasets.

```python
import bokeh
bokeh.sampledata.download()
```

Now from above files we can see that some companies' data are also available. Such as, Apple(as AAPL) , Google ( as GOOG), FB(as Facebook) , IBM, , MSFT(as Microsoft) . Now we need to visualise the stocks of these organizations.

```python
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.stocks import AAPL, IBM, FB, GOOG, MSFT
output_file('stock.html')
graph = figure(x_axis_type= 'datetime', title = 'stock closing prices')

graph.xaxis.axis_label= 'date'
graph.yaxis.axis_label= 'price in (USD)'


# plotting the line graph for Apple
x_axis_coordinates = np.array(AAPL['date'], dtype= np.datetime64) 
y_axis_coordinates = AAPL['adj_close']
color = 'blue'
legend_label= 'AAPL'
graph.line(x_axis_coordinates, y_axis_coordinates, color = color, legend_label= legend_label)


# plotting the line graph for IBM
x_axis_coordinates = np.array(IBM['date'], dtype= np.datetime64) 
y_axis_coordinates = IBM['adj_close']
color = 'red'
legend_label= 'IBM'
graph.line(x_axis_coordinates, y_axis_coordinates, color = color, legend_label= legend_label)

# plotting the line graph for Facebook
x_axis_coordinates = np.array(FB['date'], dtype= np.datetime64) 
y_axis_coordinates = FB['adj_close']
color = 'green'
legend_label= 'FB'
graph.line(x_axis_coordinates, y_axis_coordinates, color = color, legend_label= legend_label)

# plotting the line graph for Google
x_axis_coordinates = np.array(GOOG['date'], dtype= np.datetime64) 
y_axis_coordinates = GOOG['adj_close']
color = 'black'
legend_label= 'GOOG'
graph.line(x_axis_coordinates, y_axis_coordinates, color = color, legend_label= legend_label)

# plotting the line graph for Microsoft
x_axis_coordinates = np.array(MSFT['date'], dtype= np.datetime64) 
y_axis_coordinates = MSFT['adj_close']
color = 'yellow'
legend_label= 'MSFT'
graph.line(x_axis_coordinates, y_axis_coordinates, color = color, legend_label= legend_label)


#relocating the legend table to avoid abstruction of the graph

graph.legend.location= 'top_left'
show(graph)
```

Q4. Why is it essential to print a legend on a stock market chart?

It is essential to print a legend on a stock market chart because by default, each plot of a Stock chart has a legend, its items representing the series on the plot. In addition, the legend displays information about the points that are currently hovered over or, if none are hovered over, about the last points shown on the plot.
The text of a legend item includes the name of a series and, depending on the series type, the value or values of the current or last point. The legend title, enabled by default, shows the current or last date (X-value). So, enabling and disabling in any stock market chart with the help of legend() method. In any stock, the legend is enabled by default. To disable or enable it, pass false / true to the legend() method of the plot or to the enabled() method of the legend

Q5. What is the best way to limit the length of a pandas data frame to less than a year?

The best way to limit the length of a Pandas dataframe to less than a year by using DataFrame.truncate() function. Below example shows some entries before and after the passed labels of a given dataframe by using DataFrame.truncate() function.

```python
try:
    import pandas as pd
    df = pd.DataFrame({'name':['Subhash', "Shivaji", 'Malavika', 'Shakuntala', 'Sambaji', 'Bhagat'],
                       'age':[43, 36, 108, 32, 1008, 75],
                       'sex':['M','M','F','F','M','M']})
    Index = pd.date_range('2022-03-01 9:00', periods = 6, freq = 'H')
    
    df.index = Index
    
    print(df)
    
    resultant = df.truncate(before= '2022-03-01 9:00', after = '2022-03-01 12:00') # applying truncate() function
    
    print(resultant)
except:
    print('truncate function')
```

From the above output we can say that the first output is the dataframe we have created and the next output is after the implementaion of truncate() function which successfully truncated the entries before and after the passed lables in the given dataframe.

Q6. What is the definition of a 180-day moving average?

A moving average is an arithmetic mean of a certain number of data points. or we can say that Moving average refers to a series of averages of fixed size subsets of the total set of observations. It is also known as rolling average, running average, rolling means. And a 180-day moving average is the number of time periods used in the calculation. the 180-day moving average is calculated by summing the past 180 days and dividing the result by 180. We can calculate moving average with the help of Pnadas dataframe. Let's see thorugh an example as given below.

```python
try: 
    import pandas as pd
    arr = [2,4,6,8,10]
    window_size = 2
    nums_series=pd.Series(arr)
    windows = nums_series.rolling(window_size)
    moving_avgs= windows.mean()
    moving_avgs_list = moving_avgs.tolist()
    final_countdown= moving_avgs_list[window_size -1:]
    print(final_countdown)
except:
    print('moving average calculation')
    
```
