4.Examine the various forms of machine learning.

Mainly, there are four types of machine learning.

i. Supervised Learning : This machine learning is based on supervision. In this machine learning technique, we train the machine using the labelled dataset and on the basis of the training the machine predicts the result. Briefly , first and foremeost we train the machine with input and corresponding output and then we ask the machine to predict the output using the test dataset.

ii. Unsupervised Learning : Unsupervised learning technique is different from supervised and supervision is not involved in unsupervised learning. In unsupervised machine learning, the machine is trained using the unlabeled dataset, and the machine predicts the output without any supervision. In unsupervised learning, the models are trained with the data that is neither classified nor labelled, and the model acts on that data without any supervision.

iii. Semi-Supervised Machine Learning : This machine learning algorithm lies in between Supervised and Unsupervised machine learning. It represents the intermediate ground between Supervised (With Labelled training data) and Unsupervised learning (with no labelled training data) algorithms and uses the combination of labelled and unlabeled datasets during the training period. Although Semi-supervised learning is the middle ground between supervised and unsupervised learning and operates on the data that consists of a few labels, it mostly consists of unlabeled data. As labels are costly, but for corporate purposes, they may have few labels. 

iv. Reinforcement Learning : Reinforcement learning works on a feedback-based process, in which an AI agent (A software component) automatically explore its surrounding by hitting & trail, taking action, learning from experiences, and improving its performance. Agent gets rewarded for each good action and get punished for each bad action; hence the goal of reinforcement learning agent is to maximize the rewards. In reinforcement learning, there is no labelled data like supervised learning, and agents learn from their experiences only.


5.Can you explain what a well-posed learning problem is? Explain the main characteristics that must
be present to identify a learning problem properly.

A (machine learning) problem is said to be a well-posed if a solution to it exists, if that solution is unique, and if that solution depends on the data / experience but it is not sensitive to changes in the data / experience.
A computer program is said to learn from experience E in context to some task T and some performance measure P, if its performance on T, as was measured by P, upgrades with experience E. 
Some examples of well-posed learning problem are:

To filter emails as spam or not in a better way, face recognition problem, Robot driving driving problem, Automatic translation fo documents.

The main characteristics that must be present to identify a learning problem properly are :
The class of tasks (T)
The measure of performance to be improved (P)
The source of experience (E)

8.Can you explain the various forms of supervised learning? Explain each one with an example
application.

The various forms of supervised learning are :

i. Regression: In regression, a single output value is produced using training data. This value is a probabilistic interpretation, which is ascertained after considering the strength of correlation among the input variables. For example, regression can help predict the price of a house based on its locality, size.

ii. Classification: It involves grouping the data into classes.  For example, if we are thinking of extending credit to a person, we can use classification to determine whether a person would be a loan defaulter or not. 

iii. Naive Bayesian Model : The Bayesian model of classification is used for large finite datasets. It is a method of assigning class labels using a direct acyclic graph. The graph comprises one parent node and multiple children nodes. And each child node is assumed to be independent and separate from the parent. 
One example is Decision Trees : A decision tree is a flowchart-like model that contains conditional control statements, comprising decisions and their probable consequences. The output relates to the labelling of unforeseen data. A decision tree can be used to solve problems with discrete attributes as well as boolean functions. Some of the notable decision tree algorithms are ID3 and CART.

iv. Random Forest Model: The random forest model is an ensemble method. It operates by constructing a multitude of decision trees and outputs a classification of the individual trees. Suppose we want to predict which undergraduate students will perform well in GMAT. A random forest model would accomplish the task, given the demographic and educational factors of a set of students who have previously taken the test for graduate programmes. 

v.  Neural Networks :This algorithm is designed to cluster raw input, recognize patterns, or interpret sensory data. Neural network is used in image recognition.

vi.Support Vector Machine: (SVM) is a supervised learning algorithm developed in the year 1990. It draws from the statistical learning theory developed by Vap Nick. Some examples are bioinformatics, pattern recognition, and multimedia information retrieval.

Suppose one already knows from their previous work (or experience) that, the shape of each and every fruit present in the basket so, it is easy for them to arrange the same type of fruits in one place.

This is because it has a response variable which says that feature will differ from fruit to fruit that meanms ecah and every fruit has got a particular feature.This type of information is deciphered from the data that is used to train the model.This type of learning is Supervised Learning.

Unsupervised Learning: Let's take the same example in case of Unsupservised Learning. Suppose there is a basket and it is filled with some fresh fruits. The task is to arrange the same type of fruits at one place.
This time there is no information about those fruits beforehand, its the first time that the fruits are being seen or discovered . So how to group similar fruits without any prior knowledge about those.
First, any physical characteristic of a particular fruit is selected. Suppose the colour of those fruits are selected.

Then the fruits are arranged on the basis of the colour. The groups will be something as shown below:
RED COLOR GROUP: apples & cherry fruits.
GREEN COLOR GROUP: avocados & grapes.

The job has been accomplished and the fruits are sorted on the basis of colours and sizes.
Here, there is no need to know or learn anything beforehand. That means, no train data and no response variable is needed and this type of learning is Unsupervised Learning.

It is a part of Supervised Machine learning. Linear regression is applied in case of stockprice prediction, house price or real estate price prediction or height-weight prediction.Linear Regression shows the linear relationship between dependent variable (y) and one or more independent variable (x). Since it shows the linear relationship between independent and dependent variable then the value of the dependent variable changes in accordance with the value of independent variable. The linear regression model provides a sloped straight line representing the relationship between the variables ( dependent and independent). The mathematical equationof linear regression is :
y = a0 + a1x + €

where, y = dependent variable, x = independent variable, a0 = intercept of the line, a1 = linear regression coefficient and € = random error.

Linear Regression can be of two types. 

Simple Linear Regression:
If a single independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Simple Linear Regression.
Multiple Linear regression:
If more than one independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Multiple Linear Regression.

Linear Regression Line
A linear line showing the relationship between the dependent and independent variables is called a regression line. A regression line can show two types of relationship:

Positive Linear Relationship:
If the dependent variable increases on the Y-axis and independent variable increases on X-axis, then such a relationship is termed as a Positive linear relationship.
The equation of positive linear relationship is : y = a0 + a1x

Negative Linear Relationship:
If the dependent variable decreases on the Y-axis and independent variable increases on the X-axis, then such a relationship is called a negative linear relationship.
The equation of negative linear relationship is : y = -a0 + a1x
