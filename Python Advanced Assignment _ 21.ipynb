Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is
one? Why are the latter considered “good enough”?

The distinction between true random number generators (TRNGs) and pseudo-random number generators (PRNGs) is that TRNGs use an unpredictable physical means to generate numbers (like atmospheric noise), and PRNGs use mathematical algorithms (completely computer-generated or generated by software).

Pseudo-random numbers are generated by computer software which runs on regular hardware. and is highly deterministic. That means it runs the same everytime. It is really difficult to generate a real random number in software as it runs also predictably to be considered random.That is why the properly implemented pseudo-random numbers are good enough for most of the applications.

Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

 The two main factors that influence the behaviour of a "normal" probability distribution are the mean and the standard deviation. The parameters or factors determine the shape and probabilities of the distribution. The shape of the distribution changes as the parameter values change.
 For a normal distribution, 68% of the observations are within +/- one standard deviation of the mean, 95% are within +/- two standard deviations, and 99.7% are within +- three standard deviations.

Q5. In the short term, how can you expect a probability distribution to behave? What do you think will
happen as the number of trials grows?

In a probability distribution , the weighted average of possible values of a random variable, with weights given by their respective theoretical probabilities, is known as the expected value , usually represented by E(x) .
The expected value informs about what to expect in an experiment "in the long run", after many trials. In most of the cases, there could be no such value in the sample space.
The weighted average formula for expected value is given by multiplying each possible value for the random variable by the probability that the random variable takes that value, and summing all these products. It can be written as
E(x)=∑xiP(xi) ,
where xi covers all possible values for the random variable, and P(xi) is the respective theoretical probability.

E(x) is also named the mean of the probability distribution and it tells what to expect as the number of trial grows.

Q8. What is the relationship between exponentiation and logarithms?

Logarithmic functions are the inverses of exponential functions. The inverse of the exponential function y = a^b is b = a^y. The logarithmic function y = loga(b) ( y is equal to log b to the base a) is defined to be equivalent to the exponential equation b = a^y.

Q9. What are the three logarithmic functions that Python supports?

The three logarithmic functions that Python supports are: log2(x), log(x, Base), log10(x).

In order to use the functionalities of Log functions, we need to import the math module using the below statement.
import math
We all need to take note of the fact that the Python Log functions cannot be accessed directly. We need to use the math module to access the log functions in the code.
Syntax:
math.log(x)

log2(x): The math.log2(x) function is used to calculate the logarithmic value of a numeric expression of base 2.
Syntax:
math.log2(numeric expression)

log(x,Base): The math.log(x,Base) function calculates the logarithmic value of x i.e. numeric expression for a particular (desired) base value.
Syntax:
math.log(numeric_expression,base_value)

log10(x): The math.log10(x) function calculates the logarithmic value of the numeric expression to the base 10.
Syntax:
math.log10(numeric_expression)
 Some example have been set up below with respect base 2, 10 and the given base.

```python
# Example of log2(x)
try:
    import math
    print("Log value for base 2 is : ")
    print(math.log2(16))
except:
    print('example of log2(x)')
```

```python
# Example of log10(x)
try:
    import math
    print("The log value for base 10 is : ")
    print(math.log10(43))
except:
    print("The example of log10(x)")
```
